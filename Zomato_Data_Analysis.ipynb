***Zomato Data Analysis Project***






**Importing Libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**Create the Dataframe**

```python

df=pd.read_csv('/content/Zomato data .csv')
```

```python
df.head()
```

```python
df['rate']=df['rate'].str.split('/').str[0]
```

```python
sns.countplot(x=df['listed_in(type)'])
plt.xlabel('Types of Resuaurant')

```

**Conclusion: Majority of the restaurant falls in Dinning category.**

```python
df.head()
```

```python
group_data=df.groupby('listed_in(type)')['votes'].sum()
result=pd.DataFrame({'votes':group_data})
plt.plot(result,c='green', marker='o')
plt.xlabel('Type of Restaurant', c='red', size=20)
plt.ylabel('Votes', c='red', size=20)
```

***Conclusion: Dining resuaurants has received maximum votes***

```python
plt.hist(df['rate'], bins=5)
plt.xlabel('Ratings Distribution')
plt.show()
```

***Conclusion: The majority resuaurant receives ratings from 3.6 to 4.1 ***

```python
df.head()
```

```python
sns.countplot(x=df['approx_cost(for two people)'])
```

***Conclusion: The majority of couples prefer restaurants with an approximate cost of 300 rupees.***

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='online_order',y='rate', data=df)
```

***Conclusion: Offline order received lower rating in comparision to online order***

```python
pivot_table=df.pivot_table(index='listed_in(type)', columns='online_order', aggfunc='size', fill_value=0)
sns.heatmap(pivot_table, annot=True, cmap='YlGnBu', fmt='g')
plt.title('Online Order Distribution by Restaurant Type')
plt.xlabel('Online Order')
plt.ylabel('Restaurant Type')
plt.show()
```

***Conclusion: Dining Restaurants primarily accept offline orders wheres cafes primarily receive online orders. This suggests that clients prefer orders in person at restaurants, but prefer online ordering at cafes.***
