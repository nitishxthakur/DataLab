```python
death_over=data['over']>15
x=data[death_over]
x
```

```python
batsman_list=x.groupby('batsman')['batsman_runs'].count()>200
new_list=batsman_list[batsman_list].index.tolist()
```

```python
final=data[data['batsman'].isin(new_list)]
```

```python
runs=final.groupby('batsman')['batsman_runs'].sum();
balls=final.groupby('batsman')['batsman_runs'].count();
```

```python
sr=(runs/balls)*100;


```
