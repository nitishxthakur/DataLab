***Data Analysis and Visualization***

```python
cancelled_perc=df['is_canceled'].value_counts(normalize=True)
cancelled_perc
```

```python
plt.figure(figsize=(10,6))
plt.title('Reservation Status Count')
plt.bar(['Not Canceled','Canceled'],df['is_canceled'].value_counts())
plt.show()
```

```python
plt.figure(figsize=(8,4))
ax1=sns.countplot(x='hotel',hue='is_canceled', data=df, palette='Blues')
legend_labels,_=ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor=(1,1))
plt.title('Reservation status in different hotels')
plt.xlabel('Hotel')
plt.ylabel('Number of Reservations')
plt.show()
```

```python
resort_hotel=df[df['hotel']=='Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize=True)
```

```python
city_hotel=df[df['hotel']=='City Hotel']
city_hotel['is_canceled'].value_counts(normalize=True)
```

```python
resort_hotel=resort_hotel.groupby('reservation_status_date')[['adr']].mean()
mean_hotel=city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize=(20,8))
plt.title('Average Daily Rate in City Hotel and Resort Hotel', fontsize=30)
plt.plot(resort_hotel.index, resort_hotel['adr'], label='Resort Hotel')
plt.plot(city_hotel.index, city_hotel['adr'], label='City Hotel')
plt.legend(fontsize=20)
plt.show()
```

```python
df['month']=df['reservation_status_date'].dt.month
plt.figure(figsize=(10,6))
ax1=sns.countplot(x='month', hue='is_canceled',data=df)
```

```python
plt.figure(figsize=(15,8))
plt.title('ADR per Month', fontsize=30)
# Pass 'month' and 'adr' as column names within the 'data' argument
sns.barplot(x='month', y='adr', data=df[df['is_canceled']==1].groupby('month')[['adr']].sum().reset_index())
plt.show()
```

```python
cancled_data=df[df['is_canceled']==1]
top_10_country=cancled_data['country'].value_counts()[:10]
plt.figure(figsize=(8,8))
plt.title('Top 10 countries with reservation canceled')
plt.pie(top_10_country, autopct='%.2f', labels=top_10_country.index)
plt.show()
```

```python
df['market_segment'].value_counts()
```

```python
cancled_data['market_segment'].value_counts(normalize=True)
```

```python
cancled_df_adr=cancled_data.groupby('reservation_status_date')[['adr']].mean()
cancled_df_adr.reset_index(inplace=True)
cancled_df_adr.sort_values('reservation_status_date', inplace=True)

not_canceled_data=df[df['is_canceled']==0]
not_cancled_df_adr=not_canceled_data.groupby('reservation_status_date')[['adr']].mean()
not_cancled_df_adr.reset_index(inplace=True)
not_cancled_df_adr.sort_values('reservation_status_date', inplace=True)

plt.figure(figsize=(20,6))
plt.title('Average Daily Rate')
plt.plot(not_cancled_df_adr['reservation_status_date'], not_cancled_df_adr['adr'], label='Not Canceled')
plt.plot(cancled_df_adr['reservation_status_date'], cancled_df_adr['adr'], label='Canceled')
plt.legend()
plt.show()
```
